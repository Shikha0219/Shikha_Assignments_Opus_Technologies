1.	What is Regression?

```python
'''Regression in machine learning is a supervised learning technique used to predict continuous numerical values. It works by learning the relationship between input variables (features) and an output variable (target).there is one dependent variable, which is the value we want to predict, and one or more independent variables, which are the factors that influence the prediction. The model studies past or training data to understand how these variables are related.The model can estimate or predict future values based on new input data.'''
```

2.	What is Linear Regression?

```python
'''Linear Regression is a type of regression technique in machine learning used to predict a continuous numerical value by finding a linear relationship between the input variable(s) and the output variable.It assumes that the relationship between the independent variable (feature) and the dependent variable (target) can be represented by a straight line. The model tries to find the best-fit line that shows how the variables are related.
The equation of linear regression  y = mx + b'''
```

3.	When to use Linear Regression? Explain the equation of a straight line.

```python
'''Linear Regression is used when we want to predict a continuous numerical value and when there is a linear relationship between the input variables and the output variable. It works well when the change in the dependent variable is proportional to the change in the independent variable.
It is commonly used in situations such as:
Predicting house prices based on size or location
Predicting sales or revenue using past sales data
Equation of a Straight Line
The equation of a straight line used in Linear Regression is:  y=mx+b
y = Dependent variable (predicted value)
x = Independent variable (input feature)
m = Slope of the line, which shows how much y changes when x increases by one unit
b = Intercept, which is the value of y when x = 0
This equation represents the best-fit line.'''
```

9.	What is multicollinearity?

```python
'''Multicollinearity occurs in regression when two or more independent variables are highly correlated with each other.
This means that the variables contain similar information about the same relationship which makes it difficult for the model to determine the individual effect of each variable on the dependent variable.'''
```

10.	How to detect multicollinearity?

```python
'''Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. This makes it difficult to determine the individual effect of each variable on the dependent variable and can lead to unstable regression coefficients.
Correlation Matrix
A correlation matrix shows the correlation coefficients between all pairs of independent variables. If the correlation value between two variables is very high
Variance Inflation Factor (VIF)
Variance Inflation Factor is one of the most common methods used to detect multicollinearity. It measures how much the variance of a regression coefficient increases due to correlation among predictors.'''
```

11.	What are the remedies for multicollinearity?

```python
'''Remove Highly Correlated Variables
One of the simplest solutions is to remove one of the variables that are highly correlated. If two predictors contain similar information, keeping only one of them can reduce multicollinearity.
Combine Variables
Highly correlated variables can sometimes be combined into a single variable. For example, related features may be combined by taking their average or creating an index variable.
 Apply Regularization Techniques
Regularization methods such as Ridge Regression or Elastic Net can help handle multicollinearity by adding penalty terms that shrink the regression coefficients and stabilize the model.
Use Principal Component Analysis (PCA)
Principal Component Analysis transforms correlated variables into a smaller set of uncorrelated components. These components can then be used as predictors in the regression model.'''
```

12.	What is the R-Squared Statistics?

```python
'''R-Squared Statistic is a measure used in regression to evaluate how well the model explains the variability of the dependent variable.
It shows the proportion of the variance in the target variable that is explained by the independent variables in the model.'''
```

13.	What is an adjusted R-Squared Statistics?

```python
'''Adjusted R-Squared is an improved version of R-Squared used in regression models. It measures the independent variables explain the variation in the dependent variable, but it also considers the number of variables used in the model. Adjusted R-Squared adjusts the value of R-Squared so that adding unnecessary variables does not falsely improve the model performance.'''
```

14.	Why do we use adj R-squared?

```python
'''Adjusted R-Squared because R-Squared can sometimes give misleading results when more variables are added to a model.
the R-Squared value always increases when we add new independent variables, even if those variables do not actually improve the model. Adjusted R-Squared is only increases when a new variable actually improves the model performance.'''
```

15.	Why adj R-squared decreases when we use incompetent variables?



```python
'''Adjusted R-Squared considers both the fit of the model and the number of independent variables used. When a variable that does not contribute useful information is added the model becomes more complex but the improvement in prediction is very small or does not exist.Adjusted R-Squared penalizes the model for adding unnecessary variables its value decreases when irrelevant variables are included.'''
```

19.	What are Bias and Variance? What is Bias Variance Trade-off?

```python
'''Bias is the error that occurs when a model is too simple and cannot capture the real pattern in the data. A high bias model makes strong assumptions about the data and may lead to underfitting
Variance is the error that occurs when a model is too complex and learns the training data too closely. This can lead to overfitting, where the model performs well on training data but poorly on new data.The bias–variance trade-off is the balance between bias and variance when building a machine learning model. If the model is too simple, it will have high bias and low variance. If the model is too complex, it will have low bias but high variance.
The goal is to find a model that balances both bias and variance so that it performs well on both training data and new data.'''
```

25.	What is regularization, and why do we need it?

```python
'''the objective is usually to minimize the error between predicted and actual values. Regularization modifies this objective by adding a constraint that discourages very large coefficient values. This helps the model remain simpler and more generalizable
Prevents Overfitting
Regularization reduces the chance of the model fitting noise in the training data. By limiting the size of coefficients, it forces the model to focus on the most important patterns
Handles Multicollinearity
When independent variables are highly correlated, regularization helps stabilize the regression coefficients and improves model performance.'''
```

26.	Explain Ridge Regression.

```python
'''Ridge Regression is a type of linear regression technique used to handle overfitting and multicollinearity in machine learning models. It is a regularization method that modifies the ordinary least squares (OLS) regression by adding a penalty term to the loss function. This penalty discourages large values of regression coefficients and helps make the model more stable.
the objective is to minimize the sum of squared errors between the predicted values and the actual values. Ridge regression improves this by adding the squared magnitude of the coefficients as a penalty.'''
```

27.	Explain Lasso Regression.

```python
'''Lasso Regression (Least Absolute Shrinkage and Selection Operator) is a type of linear regression that uses L1 regularization to improve the performance of a model. It is mainly used to prevent overfitting and to perform feature selection by shrinking some regression coefficients to zero.the model minimizes the sum of squared errors between the actual values and predicted values. Lasso regression modifies this objective by adding a penalty equal to the absolute value of the coefficients.'''
```

28.	Explain Elastic Net.

```python
'''Elastic Net is a regularization technique used in machine learning that combines the properties of Ridge Regression (L2 regularization) and Lasso Regression (L1 regularization). It is mainly used to improve model performance, prevent overfitting, and handle situations where there are many correlated features. the goal is to minimize the difference between actual values and predicted values. Elastic Net modifies the objective function by adding both L1 and L2 penalty terms to the loss function.'''
```
